## 2. Feature Distributions
Density plots were created to visualize the distribution of each feature.
Setosa shows clear separability in some features, especially in petal length and width, where its distribution does not overlap with those of Versicolor and Virginica.
Versicolor and Virginica exhibit overlapping distributions, especcialy in sepal width and length, indicating potential challenges on separating these two classes.

```python
X = iris.data
y = iris.target
feature_names = iris.feature_names
target_names = iris.target_names
palette = sns.color_palette("Set1", n_colors=len(target_names))
fig, axes = plt.subplots(2, 2, figsize=(12, 10))
axes = axes.ravel()

for i in range(X.shape[1]):
    ax = axes[i]
    
    for class_label in np.unique(y):
        class_color = palette[class_label]
        sns.kdeplot(X[y == class_label, i], ax=ax, label=target_names[class_label], fill=True, color=class_color)

    ax.set_xlabel(feature_names[i][:-5])
    ax.set_ylabel('Density')
    ax.set_title(f'Distribution of {feature_names[i][:-5]} by class')
    ax.legend()
```

## *3. Feature Correlation
